```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.mega.pk/mobiles/1/'
response = requests.get(url)
response.status_code
```

```python
name=[]
mobile_price=[]
Mobile_detail = []
for num in range(1,7):
    url = f'https://www.mega.pk/mobiles/{num}/'
    response = requests.get(url)
    soup=BeautifulSoup(response.text,'lxml')

    titles=soup.find_all('div',id="lap_name_div")
    detailer=soup.find_all('ul',class_="detailer")
    cat_price=soup.find_all('div',class_="cat_price")


   
    for title in titles:
        mobile=title.find('h3').text.strip()
        name.append(mobile.upper())
    for detail in detailer:
        dic={}
        for des in detail.find_all('li'):
            li_tag = des.text
            lst = li_tag.split(':')
            if len(lst)>1:
                dic[lst[0].strip('- ')]=lst[1].strip()
            else:
                if 'No 5G Support' not in lst[0].strip('- ') and 'No Finger Print' not in lst[0].strip('- '):
                    dic[lst[0].strip('- ')]='Yes'
                  
        Mobile_detail.append(dic)  
        #print(dic)
        #print('--------------------------------')
       
    for price in cat_price:
        mobile_price.append(price.text.strip())
    
```

```python
print(len(name))
print(len(mobile_price))
print(len(Mobile_detail))
```

```python
import pandas as pd
pd.set_option('max_colwidth' , None)
```

```python
df = pd.DataFrame({'Title':name,
                   'Detail':Mobile_detail,
                   'Price':mobile_price,}
            )
```

```python
df.tail()
```

```python
df.insert(0,'Brand',df['Title'].str.split(' ',n=1).str[0])
```

```python
df.insert(df.columns.get_loc('Brand')+1,
          'Model',
          df['Title'].str.extract('^(.*?)\s*(\d+GB|PRO|NEW)' )[0]
         )
```

```python
df.insert(df.columns.get_loc('Title')+1,
          'RAM',
          df['Detail'].apply(lambda x: x['RAM'] if 'RAM' in x else 'N/A'  )
         )
```

```python
df.insert(df.columns.get_loc('RAM')+1,
          'Storage',
          df['Detail'].apply(lambda x: x['Storage'] if 'Storage' in x else 'N/A'  )
         )
```

```python
df.insert(df.columns.get_loc('Storage')+1,
          'Camera',
          df['Detail'].apply(lambda x: x['Camera'] if 'Camera' in x else 'N/A'  )
         )
```

```python
df.insert(df.columns.get_loc('Camera')+1,
          'Battery',
          df['Detail'].apply(lambda x: x['Battery'] if 'Battery' in x else 'N/A'  )
         )
```

```python
df.insert(df.columns.get_loc('Battery')+1,
          'Screen',
          df['Detail'].apply(lambda x: x['Screen'] if 'Screen' in x else 'N/A'  )
         )
```

```python
df.insert(df.columns.get_loc('Screen')+1,
          '5G Support',
          df['Detail'].apply(lambda x: x['5G Support'] if '5G Support' in x else 'No'  )
         )
```

```python
df.insert(df.columns.get_loc('5G Support')+1,
          'Finger Print',
          df['Detail'].apply(lambda x: x['Finger Print'] if 'Finger Print' in x else 'No'  )
         )
```

```python
df.head()
```

```python
df['Title']=df['Title'].str.replace('NON PTA','NON_PTA' )
```

```python
df.insert(df.columns.get_loc('Finger Print')+1,
          'PTA Approved', 'N/A')
```

```python
df.loc[df['Title'].str.contains('NON_PTA'),'PTA Approved']='No'
```

```python
df.loc[df['Title'].str.contains('PTA APPROVED'),'PTA Approved']='Yes'
```

```python
df.drop(columns=['Title', 'Detail'], inplace=True)
```
